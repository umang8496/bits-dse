```python
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_hub as hub

from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
print("Tensorflow Version: ", tf.__version__)
print("Eager mode: ", tf.executing_eagerly())
print("Hub version: ", hub.__version__)
print("GPU is", "available" if tf.config.list_physical_devices('GPU') else "NOT AVAILABLE")
```

```python
df=pd.read_csv("data1_0.25.csv",header=None)
df.columns =['feature_1', 'feature_2', 'class_label']  #labelling columns
df.head()
```

```python
#Creating X and Y datasets
X= df.iloc[:,0:2]
y= df.iloc[:,2]

#Applying min-max scaling
min_max_scaler = preprocessing.MinMaxScaler()
X = min_max_scaler.fit_transform(X)
```

Use appropriate learning rate, activation, and loss functions and also mention the reason for choosing the same. Report, compare, and explain the observed accuracy and minimum loss achieved.
```python
# Spliting the X and y as 70% training and 30% validation data.
X_train, X_valid,y_train, y_valid = train_test_split(X, y, test_size=0.3)
```

```python
print(X_train.shape)
X_train
```

#### Model 1 
    Activation used : sigmoid 
    Batch size used : 528 
    Epochs          : 50 
    Hidden layers   : 2 hidden layers of 12 and 4 neurons

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.InputLayer(input_shape=[2]))
model.add(tf.keras.layers.Dense(6,activation='sigmoid'))
model.add(tf.keras.layers.Dense(4,activation='sigmoid'))
model.add(tf.keras.layers.Dense(1))

model.summary()
```

```python
from tensorflow.keras.optimizers import SGD
opt = SGD(lr=0.01)
model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train,
                    y_train,
                    epochs=100,
                    batch_size=528,
                    validation_data=(X_valid, y_valid),
                    verbose=1)
```

```python
model2 = tf.keras.Sequential()
model2.add(tf.keras.layers.InputLayer(input_shape=[2]))
model2.add(tf.keras.layers.Dense(4, activation='relu'))
model2.add(tf.keras.layers.Dense(1))
model2.summary()
```

```python
model2.compile(optimizer='adam',
              loss=tf.losses.BinaryCrossentropy(from_logits=True),
              metrics=[tf.metrics.BinaryAccuracy(threshold=0.0, name='accuracy')])
```

```python
history = model2.fit(X_train,
                    y_train,
                    epochs=50,
                    batch_size=528,
                    validation_data=(X_valid, y_valid),
                    verbose=1)
```
